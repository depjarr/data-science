### Line plot

```python
# sample data

x = [1,2,3,4,5]
y = [10,25,18,30,15]

plt.plot(x,y)
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Line plot sederhana")
plt.show()
```

### Scatter Plot

```python
# sample data

x = [1,2,3,4,5]
y = [10,25,18,30,15]

plt.scatter(x,y)
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Scatter plot sederhana")
plt.show()
```

### Bar chart

```python
# perbandingan kategori

kategori = ["A", "B", "C", "D"]
nilai = [10,20,15,25]

plt.bar(kategori, nilai, color ='red')
plt.xlabel("Kategori")
plt.ylabel("Nilai")
plt.title("Bar chart sederhana")
plt.show()
```

### Histogram

```python
# persebaran data
data = np.random.randn(1000)

plt.hist(data, color='blue', bins=40)
plt.xlabel("Nilai")
plt.ylabel("Frekuensi data")
plt.title("Distribusi data")
plt.show()
```

### Pie Chart

```python
labels = ['A', 'B', 'C', 'D']
sizes = [10,30,40,15]

plt.pie(sizes,labels=labels, autopct='%1.1f%%')
plt.show()
```

## Statistik

```python
import numpy as np
import pandas as pd
```

```python
# Dataset umur pegawai

ages = [25,30,35,40,45,50,55,65]

# hitung rerata
rata_usia = np.mean(ages)
print(f"Rata-rata dari usia karyawan : {rata_usia}")

# hitung median
med_usia = np.median(ages)
print(f"Median dari usia karyawan : {med_usia}")

# hitung varians
var_usia = np.var(ages)
print(f"Varians dari usia karyawan : {var_usia}")
```

## Descriptive statistic case

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# contoh pendapatan pelanggan
data = {
    'Pendapatan' : [100,150,200,250,300,350,400,1500]

}
df = pd.DataFrame(data)
```

```python
# menghitung statistik

mean_val = df['Pendapatan'].mean()
median_val = df['Pendapatan'].median()
variance_val = df['Pendapatan'].var()
std_val = df['Pendapatan'].std()

print(f"Rata-rata pendapatan : {mean_val}")
print(f"Median pendapatan : {median_val}")
print(f"Variance pendapatan : {variance_val}")
print(f"STD pendapatan : {std_val}")
```

```python
# histogram
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt. hist(df['Pendapatan'], bins = 10, color='skyblue', edgecolor='black')



plt.xlabel("Pendapatan")
plt.ylabel("Frekuensi")
plt.title("Histogram Pendapatan")
plt.show()

# boxplot
plt.subplot(1,2,1)
plt.boxplot(df['Pendapatan'], vert=False)
plt.xlabel("Pendapatan")
plt.ylabel("Frekuensi")
plt.title("Boxplot Pendapatan")
plt.show()
```

terdapat satu outlier (pencilan) sebesar 1500

```python
# distribusi normal

mu, sigma = 165,10
x = np.linspace(130,200,100)

pdf = norm.pdf(x,mu,sigma)

plt.plot(x,pdf,label='N(165,10)')
plt.title("Distribusi Normal Tinggi Badan")
plt.xlabel("Tinggi (cm)")
plt.ylabel("Probabilitas")
plt.legend()
plt.show()
```
